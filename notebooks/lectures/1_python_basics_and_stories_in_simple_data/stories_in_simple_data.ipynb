## Typology of data stories

Source: https://datajournalism.com/read/handbook/one/understanding-data/data-stories

1. Measurement - X spent $Y dollars on Z last month
2. Proportion - X spent Y% of their total budget on Z
3. Internal comparison - X spend more on P than on Q
4. External comparison - X spend more on Z than J spends on M,N & Z
5. Change over type - here is how much X spent on Z over the past 5 years
6. League/rank tables - by region, organization etc.
7. Categories - Org type A vs Org type B spending
8. Association
